```python
# Define a function to list files in a given category folder
def list_files_in_category(category_folder):
    category_files = os.listdir(category_folder)
    print(f"Files in the '{os.path.basename(category_folder)}' category folder:")
    print(category_files)

# List files in each category within the TRAIN folder
for category in train_files:
    category_folder = os.path.join(train_folder, category)
    list_files_in_category(category_folder)

# List files in each category within the TEST folder
for category in test_files:
    category_folder = os.path.join(test_folder, category)
    list_files_in_category(category_folder)
```

```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Define function to display all images in a folder
def display_images_in_folder(folder):
    for file_name in os.listdir(folder):
        file_path = os.path.join(folder, file_name)
        if os.path.isfile(file_path):
            img = mpimg.imread(file_path)
            plt.imshow(img)
            plt.axis('off')
            plt.title(file_name)
            plt.show()

# Display images in TRAIN folder
print("Displaying images in TRAIN folder:")
for category in train_files:
    category_folder = os.path.join(train_folder, category)
    display_images_in_folder(category_folder)

# Display images in TEST folder
print("Displaying images in TEST folder:")
for category in test_files:
    category_folder = os.path.join(test_folder, category)
    display_images_in_folder(category_folder)
```
